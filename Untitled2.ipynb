```python
from collections import Counter 
value= input("Enter a value: ")
if value==value[::-1]:
    print("palindrome")
else:
    print("not palindrome")
    
cd=Counter(value)
for keys in sorted(cd.keys()):
    print(f'{keys} appears {cd[keys]} times');
```

```python
def fn(n):
    if n<=2:
        return n-1
    else:
        return fn(n-1)+fn(n-2)
try:
    num=int(input("enter a number: "))
    if num>0:
        print(f'fn({num})={fn(num)}')
    else:
        print("imput should be greater than 0")
        
except ValueError:
    print("try with numeric value")
    
```

```python
def bin2Dec(val):
    return int(val,2)
def oct2Hex(val):
    return int(val,8)
try:
    num1=input("enter a binary number: ")
    print(bin2Dec(num1))
    
except ValueError:
    print("invalid literal in input with base 2")
    
try:
    num1=input("enter a octal number")
    print(oct2Hex(num1))
    
except ValueError:
    print("invalid literal in input with base 8")
```

```python
import string 
sentence=input("enter a sentence: ")

wordList=sentence.strip().split(" ")
print(f' this sentence has {len(wordList)} words', end='\n\n')

digit_count=uppercase_count=lowercase_count=0

for character in sentence:
    if character in string.digits:
        digit_count+=1
    elif character in string.ascii_uppercase:
        uppercase_count+=1
    elif character in string.ascii_lowercase:
        lowercase_count+=1
print(f' this sentence has {digit_count} digits',
     f' {uppercase_count} uppercase letters',
     f' {lowercase_count} lowercase letters')
```

```python
from difflib import SequenceMatcher
str1=input("enter a string 1: ")
str2=input("enter a string 2: ")

sim= SequenceMatcher(None, str1,str2).ratio()
print("similarity btw strings\""+str1+"\"and\""+str2+"\" is:", sim)
```

```python
import matplotlib.pyplot as plt
import numpy as np
countries=['brazil','russia', 'india','china','south africa']
population=[213993437,145912025,1393409038,1444216107,61608912]
pci=[9600,11600,1300,11000,6500]
circle_size=[pop/ 1000000 for pop in population]
colors=np.arange(len(countries))
scatter=plt.scatter(population, pci, s=circle_size, c=colors, cmap='viridis',
    alpha=0.7, label='BRICS NATION')
                    
for i, country in enumerate(countries):
                    plt.annotate(country,(population[i], pci[i]), textcoords="offset points", xytext=(0,5), ha='center')
plt.colorbar(scatter, label="index")
plt.xlabel('population')
plt.ylabel('pci(USD)')
plt.title("population vs pci")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(42)
student_scores=np.random.normal(loc=70,scale=15, size=100)
plt.hist(student_scores, bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Frequency')
plt.title('distribution of student scores')
plt.show()
```
